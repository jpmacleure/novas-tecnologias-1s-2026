# Tipos de gráficos com matplotlib

### Importa bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Carrega dados de chuvas

```python
df = pd.read_csv("dados_chuvas_ficticias.csv", parse_dates=["data"])
print(df.head())
```

### Gráfico de colunas
Adequado para comparar quatidades entre diferentes categorias.

```python
contagem_chuvas
```

```python
contagem_chuvas = df["tipo_chuva"].value_counts()
# Ordenada crescente
#plt.bar(contagem_chuvas.sort_values(ascending=True).index, contagem_chuvas.sort_values(ascending=True).values)
plt.bar(contagem_chuvas.index, contagem_chuvas.values)
plt.title("Quantidade de dias por tipo de chuva")
plt.xlabel("Tipo de chuva")
plt.ylabel("Número de dias")
plt.show()
```

### Gráfico de linhas
Adequado para visualizar tendências e evoluções de sequências numérias ao longo do tempo

```python
plt.plot(df["data"], df["temperatura_media"])
plt.title("Evolução da temperatura média diária")
plt.xlabel("Data")
plt.ylabel("Temperatura °C")
plt.xticks(rotation=45)
plt.show()
```

### Gráfico de pizza/torta
Adequado para representar proporções de um todo

```python
plt.pie(contagem_chuvas.values, labels=contagem_chuvas.index, autopct="%1.1f%%")
plt.title("Distribuição dos tipos de chuva")
plt.show()
```

### Gráfico de dispersão
Adequado para visualizar correlação ou relacionamento entre duas variáveis numéricas

```python
plt.scatter(df["temperatura_media"], df["umidade"])
plt.title("Relação entre temperatura e umidade")
plt.xlabel("Temperatura °C")
plt.ylabel("Umidade %")
plt.show()
```

### Histograma
Similar ao de colunas, mas usado para mostrar distribuições de frequências

```python
plt.hist(df["temperatura_media"], bins=10)
plt.title('Distribuição da temperatura média')
plt.xlabel("Temperatura °C")
plt.ylabel("Frequência")
plt.show()
```

```python
numbers = np.random.normal(1, 100, size=10000)

plt.hist(numbers, bins=1000)
plt.title('Distribuição de números aleatórios do numpy random')
plt.show()
```
