## Separa a base de dados em 3, de acordo ao ano do registro:
- Menor que 1975 (p1)
- Maior ou igual a 1975 e menor que 2000 (p2)
- Maior ou igual a 2000 (p3)

```python
# seleciona os dados de cada período
dados_p1 = dados[dados["Ano"] < 1975]
dados_p2 = dados[((dados["Ano"] >= 1975) & (dados["Ano"] < 2000))]
dados_p3 = dados[dados["Ano"] >= 2000]
```
