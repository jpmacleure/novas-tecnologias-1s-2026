### Operadores aritméticos

```python
A = 2
B = 3

soma = A + B
subtracao = A - B
multiplicacao = A * B
divisao = A / B
potenciacao = A ** B
modulo = A % B

print(soma)
print(subtracao)
print(multiplicacao)
print(potenciacao)
print(modulo)
```

### Operadores lógicos

```python
A = 2
B = 3
print(A > B)
print(A < B)
print(A == B)
print(A != B)
```

### Estruturas condicionais

```python
idade = int(input('Digite a idade: '))
possui_habilitacao = False

if idade >= 18 and possui_habilitacao:
    print('Possui habilitação.')
elif idade >= 18:
    print('Sim')
    print('É maior de idade.')
else:
    print('Não')
    print('Não é maior de idade.')
```

### Estruturas de repetição

```python
for i in range(5):
    print(i)

cont = 0
while cont < 5:
    print(cont)
    cont += 1
```
